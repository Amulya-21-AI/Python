#### Difference between return and print

```python
#using print how output generates
def add(a, b):
    print(a + b)

result = add(2, 3)
print(result)
```

```python
#using return what happens here
def add(a, b):
    return a + b

result = add(2, 3)
print(result)
```

```python
#using both return and print
def add(a, b):
    print("Sum is:", a + b)
    return a + b

x = add(4, 6)
print("Returned value:", x)
```
