```python
# 1. Prime Number
A = 10
for i in range(2,A):
    if A%i == 0:
        print("not prime number")
        break
else:
        print("prime number")
```

```python
# 2. Fibbanacci Series
def fib(n):
    a = 0
    b = 1 
    if n <= 100:
        print(a)
    else:
        print(a)
        print(b)
        for i in range(2,n):
            c = a+b
            a = b
            b = c
            print(c)
fib(10)

#0,1,2,3,4
```
