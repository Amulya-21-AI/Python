#### Errors & Exception Handling
* Errors may occur randomly,we correct them using try,except 

```python
try:
    f = open('testfile.txt','r')
    f.write('test write this')
except IOError:
    print("Error: Could not find file")
finally:
    print("Execution completed")
```
