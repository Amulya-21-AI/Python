#### Composition - HAS a relationship
one object uses Another object

```python
class Engine:
    def start(self):
        print("Engine started")

class Car:
    def __init__(self):
        self.engine = Engine()   # Car HAS an Engine

    def run(self):
        self.engine.start()
        print("Car running")
```

```python
c = Car()
c.run()
```

#### Dynamic Extension
adding new things later while program is running.

```python
class Dog:
    def __init__(self, name):
        self.name = name

d = Dog("Tommy")

# adding new variable later
d.color = "Brown"
d.size = "Big"

print(d.name, d.color,d.size)
```

#### Polymorphism / DuckTyping
1. same function can perform on different data types
2. One thing behaving in many ways

```python
class mom:
    def hobby(self):
        print("mom likes cooking")
class amulya:
    def hobby(self):
        print("Amulya likes singing")
a = amulya()
b = mom()
a.hobby()
b.hobby()
```

#### Instance Variables
1. Variables belongs to each object seperately
2. like every object has its own copy of data - Individual data

```python
class Student:
    def __init__(self, name, marks):
        self.name = name      # instance variable
        self.marks = marks    # instance variable
        #self.age = age       can add vairiables if needed then we can increase the parameters

s1 = Student("Amulya", 85)
s2 = Student("Sushma", 90)
s3 = Student("Rupa",78)

print(s1.name, s1.marks)
print(s2.name, s2.marks)
print(s3.name,s3.marks)
```

#### Class Variables
* Class variables are shared by all objects of the class. - Common Data

```python
class Student:
    college = "Vignan College"   # class variable
    class_name = "Ms"

    def __init__(self, name):
        self.name = name

s1 = Student("Amulya")
s2 = Student("Rahul")

print(s1.college,s1.class_name)
print(s2.college)
```

#### Instance Method
Works with object data,can access instance Variables

```python
class Student:
    def __init__(self, name):
        self.name = name

    def show(self):      # instance method
        print("Student name:", self.name)

s = Student("Amulya")
s.show()
```

#### Class Method
works with class variables,it defined using @classmethod also can modify the class data

```python
class Student:
    college = "Vignan College"

    @classmethod
    def show_college(cls):
        print("College:", cls.college)

Student.show_college()
```

#### Static Method
* It is Independent Method,defined using @staticmethod
* it do not access instance or class variables


```python
class Math:
    @staticmethod
    def add(a, b):
        return a + b

print(Math.add(5, 3))
```

#### Decorators In Python
* A decorator is used to add extra functionality to function without changing its original code.
* InShort, function that modifies another function


```python
def my_decorator(func):
    def wrapper():
        print("Something before function")
        func()
        print("Something after function")
    return wrapper
```

```python
#using @ syntax
def my_decorator(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper

@my_decorator
def greet():
    print("Welcome")

greet()
```

```python
# Decorators With Arguments
def smart_divide(func):
    def wrapper(a, b):
        if b == 0:
            print("Cannot divide by zero")
        else:
            func(a, b)
    return wrapper

@smart_divide
def divide(a, b):
    print(a/b)

divide(10, 2)
divide(10, 0)
```
