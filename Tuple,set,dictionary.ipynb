### Sets in Python

##### A Set in Python is used to store multiple items with following properties.

* No duplicate elements. If try to insert same item again, it overwrites before one.
* unordered collection. When we access all items, they are accessed without any specific order and we cannot access items using indexes as we do in lists.
* Internally use hashing that makes set efficient for search, insert and delete operations.
* It gives a major advantage over a list for problems with these operations.
* Mutable, meaning we can add or remove elements after their creation,
* The individual elements within set cannot be changed directly

```python
# 1st Property
x = {"Amulya","Kurapati","india"}       #If values in set are unordered,It automatically prints in a order way
print(x)


y = {"Amulya","kurapati","amulya"}      #when searching for duplicates it also considers #case sensitves
print(y)


z = {"amulya","kurapati","amulya"}    
print(z)
```

```python
a = {"amulya", "kurapati", "india", True, 1, 2}       #System considers '1' as True, so '1' becomes duplicate
print(a)


b = {"amulya","kurapati",0,False,3}  #same like above with false, but Considers only the first Occurances
print(b)
```

```python
# length of set
a = {"amulya","kurapati",1,2}
print(len(a))
```

```python
print(a[0])      #cannot retrive with indexing
```

###### Frozen sets
* This Sets are immutable but can perform union, intersection & difference

```python
# Normal set (mutable)
s = set(["a", "b", "c"])
print("Normal Set:", s)

# Frozen set (immutable)
fs = frozenset(["e", "f", "g"])
print("Frozen Set:", fs)
```

```python
# Normal set (mutable)
s = set(["a", "b", "c"])
s[1] = 'x'
print("Normal Set:", s)

# Frozen set (immutable)
fs = frozenset(["e", "f", "g"])
print("Frozen Set:", fs)
```

### Dictionaries in Python
* Python dictionary is a data structure that stores data in key-value pairs,
* where each key is unique and is used to retrieve its associated value.
* It is mainly used when you want to store and access data by name (key) instead of by position like in list.



```python
dict1 = {"name": "Amulya", "Age": 23}
print(dict1)
```

```python
print(dict1['name'])
```

```python
# Adding/updating to the keys in dictionary
d = {1: 'amulya', 2: 'Kurapati', 3: 'India'}

# Adding a new key-value pair
d["age"] = 23

# Updating an existing value
d[1] = "Python dict"
print(d)
```

##### Removing Dictionary Items
Dictionary items can be removed using built-in deletion methods that work on keys:

* del: removes an item using its key
* pop(): removes the item with the given key and returns its value
* clear(): removes all items from the dictionary
* popitem(): removes and returns the last inserted keyâ€“value pair

```python
d = {1: 'a', 2: 'b', 3: 'c', 'age':22}

# Using del 
del d["age"]
print(d)

# Using pop() 
val = d.pop(1)
print(val)

# Using popitem()
key, val = d.popitem()
print(f"Key: {key}, Value: {val}")

# Using clear()
d.clear()
print(d)
```

#### Shallow Copy
1. shallow copy creates a new dictionary but does NOT copy inner objects completely.
2. Inner objects are still shared between original and copied dictionary.

```python
import copy

original = {"a": 1, "b": [10, 20]}
shallow = copy.copy(original)

shallow["b"][0] = 99

print(original)
```

#### Deep Copy
deep copy creates a completely independent copy.

```python
import copy

original = {"a": 1, "b": [10, 20]}
deep = copy.deepcopy(original)

deep["b"][0] = 99

print(original)
```
