## ENTRY CONTROL LOOP
Definition: Entry-Control Loop is a loop in which the condition is checked before executing the loop body.
If the condition is false initially, the loop body will not execute even once.

Python Loops that are Entry-Control:

1. for loop

2. while loop

## Range funtion in For Loop

### The range() function is commonly used with for loops to generate sequence of numbers, It can take one, two, or three arguments:

* 1-Parameter: range(stop): Generates numbers from 0 to stop-1.
* 2-Parameter: range(start, stop): Generates numbers from start to stop-1.
* 3-Parameter: range(start, stop, step): Generates numbers from start to stop-1, incrementing by step.

why only python takes n-1:
* python is Zero '0 based indexing
* python Always works with the < operator: in Range and Loops
* input which we gives to the range() should be INT, Output can be anything 'iterable'

## While Loop - works like conditional Loop

# Steps to run the While loop
1. Initializing the input Value
2. Creating while condition,to execute the loop until
3. Checks the condition
4. If satsifies
5. Print i vaue
6. If not terminates

```python
# Iterates until the condition gets satsifies/True
# Everytime it checks the condition,before executing the block of Loop
i = 1

#while i <= 5:
while i<5:
    print(i)
    i += 1

#i = 1 → condition true → print 1
#i = 2 → print 2
#i = 3 → print 3
#i = 4 → print 4
#i = 5 → print 5
#i = 6 → condition false → stop

#i=1
#while i<=5:
#print(i)
```
